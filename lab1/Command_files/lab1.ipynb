```python
import pandas as pd
import matl

data = pd.read_fwf("../Original_data/weather.txt", widths=[11,4,2,4]+[5,3]*31, header=None)
for i in range(5,66,2):
    data=data.drop(i, axis=1)

data
```

```python
melted = data.melt(id_vars =[0, 1, 2, 3])
melted.head(100)

def reindex(index):
    return round(index/2-1)


melted.variable=melted.variable.apply(reindex)
melted=melted.rename(columns={0: "id", 1: "rok", 2:"mc", 3:"co", "variable":"dzien"})
melted["data"]=melted.rok.map(str)+"-"+melted.mc.map(str)+"-"+melted.dzien.map(str)

melted=melted.drop("mc",axis=1)
melted=melted.drop("rok",axis=1)
melted=melted.drop("dzien",axis=1)
melted
```

```python
tidy=melted.pivot(index='data', columns='co', values='value')

#tidy.to_csv("../Analysis_data/weather_tidy.csv")

tidy_temp=tidy.drop("PRCP",axis=1)
tidy_prcp=tidy.drop("TMAX",axis=1).drop("TMIN",axis=1)

tidy_temp.to_csv("../Analysis_data/weather_tidy_temp.csv")
tidy_prcp.to_csv("../Analysis_data/weather_tidy_prcp.csv")
tidy
```
