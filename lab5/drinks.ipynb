```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display

pd.options.display.max_rows = None




data=pd.read_csv("Datasets/drinks.csv")
mean = data.mean()
datac=data.copy()
datac['beer_servings'] = np.where((mean["beer_servings"] <data['beer_servings']), 
                           datac['beer_servings'],     
                           "więcej")
datac['beer_servings'] = np.where((mean["beer_servings"] >data['beer_servings']), 
                           datac['beer_servings'],     
                           "mniej")
datac['spirit_servings'] = np.where((mean["spirit_servings"] <data['spirit_servings']), 
                           datac['spirit_servings'],     
                           "więcej")
datac['spirit_servings'] = np.where((mean["spirit_servings"] >data['spirit_servings']), 
                           datac['spirit_servings'],     
                           "mniej")
datac['wine_servings'] = np.where((mean["wine_servings"] <data['wine_servings']), 
                           datac['wine_servings'],     
                           "więcej")
datac['wine_servings'] = np.where((mean["wine_servings"] >data['wine_servings']), 
                           datac['wine_servings'],     
                           "mniej")
datac['total_litres_of_pure_alcohol'] = np.where((mean["total_litres_of_pure_alcohol"] <data['total_litres_of_pure_alcohol']), 
                           datac['total_litres_of_pure_alcohol'],     
                           "więcej")
datac['total_litres_of_pure_alcohol'] = np.where((mean["total_litres_of_pure_alcohol"] >data['total_litres_of_pure_alcohol']), 
                           datac['total_litres_of_pure_alcohol'],     
                           "mniej")

res = datac.where(datac["beer_servings"]=="więcej")
res2 = datac.where(datac["spirit_servings"]=="więcej")
res3 = datac.where(datac["wine_servings"]=="więcej")
res4 = datac.where(datac["total_litres_of_pure_alcohol"]=="więcej")
```
